 perform **KS-test**

```python
# ks test
ksresult = {}
for col in goodcols:
    d, p = stats.ks_2samp(df_all[col].dropna(), df_good[col])
    ksresult.update({col: {'dvalue': d, 'pvalue':p, 'same distribution?':p>0.05}})
pd.DataFrame(ksresult)
```

From the result of KS-test, we find that all pvalues are close to 1. This means we cannot reject null hypothesis. The distribution of our dataset after drop nan is almost the same as the original merged one.

perform **Z-test**

```python
# z-test
zresult = {}
for col in goodcols:
    mu = np.mean(df_all[col].dropna())
    sigma = np.std(df_all[col].dropna())
    xbar = np.mean(df_good[col])
    zvalue = (xbar - mu) / (sigma / np.sqrt(df_good[col].shape[0]))
    zresult.update({col: {'zvalue':zvalue, 'same distribution?': (zvalue>-1.96)&(zvalue<1.96) }})

pd.DataFrame(zresult)
```

The result from Z-test shows that some features fail to maintain the same mean and standard deviation as the original merged dataset.
